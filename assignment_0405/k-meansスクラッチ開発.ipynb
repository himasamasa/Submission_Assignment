```python
class ScratchKMeans():
    def __init__(self, n_clusters=2, n_init=20, n_iter=100, verbose=False):
        self.n_clusters = n_clusters
        self.n_init = n_init
        self.n_iter = n_iter
        self.verbose = verbose
        self.best_sse = 10**20
        
    def _calc_sse(self, X):
        sse=0
        for i in range(len(X)):
            for j in range(self.n_clusters):
                if j == X[i, -1]:
                    sse += np.sum((X[i,:-1] - self.myu[j])**2)
        if self.best_sse > sse:
            self.best_sse = sse
            self.record_myu = myu
            self.record_cluster = n_clusters
            
    def _update_cluster(self, X):
        for i in range(len(X)):
            dist_m = 10**20
            for j in range(self.n_clusters):
                dist = np.linalg.norm(X[i,:-1]-self.myu[j])
                if dist < dist_m:
                    dist_m = dist
                    X[i,-1] = j
                
    def _update_myu(self, X):
        for j in range(self.n_clusters):
            self.myu[j] = np.mean(X[X[:,-1]==j,:-1],axis=0)
            
    def fit(self, X):
        for i in range(self.n_init):
            self.myu = X[np.random.choice(len(X),size=self.n_clusters,replace=False)]
            cluster = np.zeros(len(X))
            X_cluster = np.concatenate([X,cluster.reshape(-1,1)],axis=1)
            for j in range(self.n_iter):
                self._calc_sse(X_cluster)
                self._update_cluster(X_cluster)
                self._update_myu(X_cluster)
                if self.verbose:
                    print("初期化：%s　学習回数：%s　最善のSSE：%s"%(i,j,self.best_sse))
                    
    def predict(self, X):
        y_pred = np.zeros(len(X))
        for i in range(len(X)):
            dist_m = 10**20
            for j in range(self.record_cluster):
                dist = np.sqrt(np.sum((X[i]-self.record_myu[j])**2))
                if dist < dist_m:
                    dist_m = dist
                    y_pred[i] = j
        return y_pred
```

# 問題9　シルエット図

```python
class ScratchKMeans_sil():
    def __init__(self, n_clusters=2, n_init=20, n_iter=100, verbose=False):
        self.n_clusters = n_clusters
        self.n_init = n_init
        self.n_iter = n_iter
        self.verbose = verbose
        self.best_sse = 10**20
        
    def _calc_sse(self, X):
        sse=0
        for i in range(len(X)):
            for j in range(self.n_clusters):
                if j == X[i, -1]:
                    sse += np.sum((X[i,:-1] - self.myu[j])**2)
        if self.best_sse > sse:
            self.best_sse = sse
            self.record_myu = myu
            self.record_cluster = n_clusters
            
    def _update_cluster(self, X):
        for i in range(len(X)):
            dist_m = 10**20
            for j in range(self.n_clusters):
                dist = np.linalg.norm(X[i,:-1]-self.myu[j])
                if dist < dist_m:
                    dist_m = dist
                    X[i,-1] = j
                
    def _update_myu(self, X):
        for j in range(self.n_clusters):
            self.myu[j] = np.mean(X[X[:,-1]==j,:-1],axis=0)
            
    def fit(self, X):
        for i in range(self.n_init):
            self.myu = X[np.random.choice(len(X),size=self.n_clusters,replace=False)]
            cluster = np.zeros(len(X))
            X_cluster = np.concatenate([X,cluster.reshape(-1,1)],axis=1)
            for j in range(self.n_iter):
                self._calc_sse(X_cluster)
                self._update_cluster(X_cluster)
                self._update_myu(X_cluster)
                self.X_cluster = X_cluster
                if self.verbose:
                    print("初期化：%s　学習回数：%s　最善のSSE：%s"%(i,j,self.best_sse))
                    
    def predict(self, X):
        y_pred = np.zeros(len(X))
        for i in range(len(X)):
            dist_m = 10**20
            for j in range(self.record_cluster):
                dist = np.sqrt(np.sum((X[i]-self.record_myu[j])**2))
                if dist < dist_m:
                    dist_m = dist
                    y_pred[i] = j
        return y_pred
```

```python
from sklearn.metrics import silhouette_samples
from matplotlib import cm

X, _ = make_blobs(n_samples=100, n_features=2, centers=4, cluster_std=0.5, shuffle=True, random_state=0)

for k in range(2,8):
    model = ScratchKMeans_sil(n_clusters=k, n_init=100, n_iter=100, verbose=False)
    model.fit(X)
    
    y_ax_lower, y_ax_upper = 0,0
    yticks = []
    
    y_km = model.X_cluster[:,-1]
    cluster_labels = np.unique(y_km)
    
    # シルエット係数を計算
    silhouette_vals = silhouette_samples(X,y_km,metric='euclidean')
    
    for i ,c in enumerate(cluster_labels):
        c_silhouette_vals = silhouette_vals[y_km == c]
        c_silhouette_vals.sort()
        y_ax_upper += len(c_silhouette_vals)
        color = cm.jet(i/ model.n_clusters)
        plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, edgecolor='none', color=color)
        yticks.append((y_ax_lower + y_ax_upper) / 2)
        y_ax_lower += len(c_silhouette_vals)
    
    silhouette_avg = np.mean(silhouette_vals)  
    plt.axvline(silhouette_avg, color='red', linestyle='--')
    plt.yticks(yticks, cluster_labels + 1)
    plt.ylabel('Cluster')
    plt.xlabel('Silhouette coefficient')
    plt.show()
```

エルボー法よりグラフが大きく曲がっている点が3であることが分かる。また、シルエット図より、同じような形と幅の図があまり見られないことから今回のクラスタ数の決定にはあまり役に立たない。
これらのことから、クラスタ数3を選択する。

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

n = 4
pca = PCA(n_components=n)
pca = pca.fit(data)
var_exp = pca.explained_variance_ratio_
cum_var_exp = np.cumsum(var_exp)
print(var_exp)
print(cum_var_exp)

fig = plt.subplots(figsize=(12,8))
plt.bar(range(1,n+1), var_exp, alpha=0.5, align='center', label='individual explained variance')
plt.step(range(1,n+1), cum_var_exp, where='mid', label='cumulative expalined variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.hlines(0.7, 0, 6,  "blue", linestyles='dashed')
plt.legend(loc='best')
plt.grid()
plt.show()
```

# 問題13　他の手法の調査

DBSCANのメリット・デメリット  
メリット  
- 最初にクラスタ数を決めなくても作成できる
- 外れ値に対してロバストである
デメリット  
- データが分からないとパラメータを決めるのが難しい
- データによりどのクラスタに属するか変わる可能性がある
- 密度に大きな違いのあるデータ集合をクラスタリングできない

t-SNEのメリット・デメリット  
メリット  
- 高次元の局所的な構造を非常によく捉える
- 大局的な構造も可能な限り捉える
デメリット  
- 内部のパラメータを変えると全く異なるクラスターが出現してしまう

LLEのメリット・デメリット  
メリット  
- 非線形多様体を学ぶことができる
デメリット  
- データに高密度の変動がある場合では失敗することがある
