# 問題5

```python
THICKNESS = 0.00008
n_ori = 43

ori_list = [THICKNESS]

folded_thickness = THICKNESS
for _ in range(n_ori):
    folded_thickness = folded_thickness * 2
    ori_list.append(folded_thickness)
    
print("厚さ： {}メートル".format(folded_thickness))
print("リストに格納された個数：{}個".format(len(ori_list)))
```

# 問題6

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
THICKNESS = 0.00008
n_ori = 43

ori_list = [THICKNESS]

folded_thickness = THICKNESS
for _ in range(n_ori):
    folded_thickness = folded_thickness * 2
    ori_list.append(folded_thickness)
    

# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(ori_list)  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

# 問題7

```python
THICKNESS = 0.00008
n_ori = 43

ori_list = [THICKNESS]

folded_thickness = THICKNESS
for _ in range(n_ori):
    folded_thickness = folded_thickness * 2
    ori_list.append(folded_thickness)
```

## その1

```python
# 線を緑にする
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(ori_list, color="green")  
plt.show()
```

## その2

```python
# 軸の値を大きく表示する。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.tick_params(labelsize=20)
plt.plot(ori_list)
plt.show()
```

## その3

```python
# 点線にする
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(ori_list, linestyle='--')
plt.show()
```
