# 問題4　目的関数

```python
def loss_func(theta,pred, y):
    lam = np.array([0,1])
    loss = np.mean(-y*np.log(pred) - (1-y)*np.log(1-pred)) + lam/2*np.mean(theta**2)
    return loss
```

```python
class ScratchLogisticRegression():
    """
    ロジスティック回帰のスクラッチ実装

    Parameters
    ----------
    num_iter : int
      イテレーション数
    lr : float
      学習率
    no_bias : bool
      バイアス項を入れない場合はTrue
    verbose : bool
      学習過程を出力する場合はTrue

    Attributes
    ----------
    self.coef_ : 次の形のndarray, shape (n_features,)
      パラメータ
    self.loss : 次の形のndarray, shape (self.iter,)
      訓練データに対する損失の記録
    self.val_loss : 次の形のndarray, shape (self.iter,)
      検証データに対する損失の記録

    """

    def __init__(self, num_iter=100, lr=0.01, C=1, bias=False, verbose=False):
        # ハイパーパラメータを属性として記録
        self.iter = num_iter
        self.lr = lr
        self.C = C
        self.lam = 1/self.C
        self.bias = bias
        self.verbose = verbose
        # 損失を記録する配列を用意
        self.loss = np.zeros(self.iter)
        self.val_loss = np.zeros(self.iter)
        
    def _hypothetical_function(self, X):
        z = X@self.theta
        pred = 1 / (1 + np.exp(-z))
        
        return pred
    
    def _gradient_descent(self, X, error):
        # tmpでi=1,2,...に正則化(lamda)を追加する。
        self.tmp = np.append(0,np.ones(X.shape[1]-1))
        grad = self.lr*(np.mean(error@X) + self.tmp*self.lam*self.theta)/len(X)
        self.theta = self.theta - grad
        return theta
    
    
    def _loss_function(self, pred, y):
        loss = np.mean(-y*np.log(pred) -(1-y)*np.log(1-pred)) + (self.lam/2)*np.mean(self.theta[1:]**2)
        
        return loss
        

    def fit(self, X, y, X_val=None, y_val=None):
        """
        ロジスティック回帰を学習する。検証データが入力された場合はそれに対する損失と精度もイテレーションごとに計算する。

        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            訓練データの特徴量
        y : 次の形のndarray, shape (n_samples, )
            訓練データの正解値
        X_val : 次の形のndarray, shape (n_samples, n_features)
            検証データの特徴量
        y_val : 次の形のndarray, shape (n_samples, )
            検証データの正解値
        """
        # ラベルの取得
        self.ylabel = np.unique(y)
        # y のラベル値を 0,1 に置き換える
        y = np.where(y==self.ylabel[0],0,1)
        print(y)
        
        if (type(y_val) != bool):
            y_val = np.where(y_val==self.ylabel[0],0,1)
        
        # bias
        if self.bias == True:
            bias_0 = np.ones(X.shape[0]).reshape(-1,1)
        else:
            bias_0 = np.zeros(X.shape[0]).reshape(-1,1)
        X = np.concatenate([bias_0, X], axis=1)
        self.theta = np.random.rand(X.shape[1])
        
        if X_val is not None:
            if self.bias == True:
                bias_0 = np.ones(X_val.shape[0]).reshape(-1,1)
            else:
                bias_0 = np.zeros(X_val.shape[0]).reshape(-1,1)
            X_val = np.concatenate([bias_0, X_val], axis=1)
        
        for i in range(self.iter):
            y_pred = self._hypothetical_function(X)
            error = y_pred - y
            self.loss[i] = self._loss_function(y_pred, y)
            if i <=10:
                print(self._loss_function(y_pred, y))
            
            if X_val is not None:
                y_val_pred = self._hypothetical_function(X_val)
                self.val_loss[i] = self._loss_function(y_val_pred, y_val)
            
            # thetaの更新    
            self._gradient_descent(X, error)
            
            if self.verbose:
                #verboseをTrueにした際は学習過程を出力
                print('{}回目の学習の損失は{},theta={}'.format(i+1,self.loss[i],self.theta))


    def predict(self, X):
        """
        ロジスティック回帰を使いラベルを推定する。

        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            サンプル

        Returns
        -------
            次の形のndarray, shape (n_samples, 1)
            ロジスティック回帰による推定結果
        """
        # bias
        if self.bias == True:
            bias_0 = np.ones(X.shape[0]).reshape(X.shape[0],1)
        else:
            bias_0 = np.zeros(X.shape[0]).reshape(X.shape[0],1)
        X = np.concatenate([bias_0, X], axis=1)
        pred = self._hypothetical_function(X)
        pred_label = np.where(pred<0.5,self.ylabel[0], self.ylabel[1])
        
        return pred_label

    def predict_proba(self, X):
        """
        ロジスティック回帰を使い確率を推定する。

        Parameters
        ----------
        X : 次の形のndarray, shape (n_samples, n_features)
            サンプル

        Returns
        -------
            次の形のndarray, shape (n_samples, 1)
            ロジスティック回帰による推定結果
        """
        # bias
        if self.bias == True:
            bias_0 = np.ones(X.shape[0]).reshape(X.shape[0],1)
        else:
            bias_0 = np.zeros(X.shape[0]).reshape(X.shape[0],1)
        X = np.concatenate([bias_0, X], axis=1)
        pred = self._hypothetical_function(X)
        
        return pred
```

# 問題5　学習と推定

```python
# irisデータセットの呼び出し
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
def scratch_train_test_split(X,y,train_size=0.8,random_state=0):
    np.random.seed(random_state)
    y = y.reshape(-1,1)
    Xy = np.concatenate([X,y],axis=1)
    size = len(Xy)
    pick = int(np.round(size*train_size))
    train_pick = np.random.choice(np.arange(size),pick,replace=False)
    test_pick = np.delete(np.arange(size),train_pick)
    train = Xy[train_pick,:]
    test = Xy[test_pick,:]
    X_train = train[:,0:(Xy.shape[1]-y.shape[1])].reshape(-1,X.shape[1])
    y_train = train[:,-y.shape[1]].reshape(-1,)
    X_test = test[:,0:(Xy.shape[1]-y.shape[1])].reshape(-1,X.shape[1])
    y_test = test[:,-y.shape[1]].reshape(-1,)
    return X_train, X_test, y_train, y_test
```

```python
# irisデータセットの呼び出し
data = load_iris().data
target = load_iris().target.reshape(-1,1)
iris = np.concatenate([data,target],axis=1)
iris = pd.DataFrame(iris)
iris_X = iris.loc[iris[4]!=0,2:3].values
iris_y = iris.loc[iris[4]!=0,4].values
X = iris_X
y = iris_y
X_train, X_test, y_train, y_test = scratch_train_test_split(X,y,train_size=0.8,random_state=0)
```

```python
print('X_train.shape:',X_train.shape)
print('y_train.shape:',y_train.shape)
print('X_test.shape:',X_test.shape)
print('y_test.shape:',y_test.shape)
```

```python
# 標準化
scaler = StandardScaler()
scaler.fit(X_train)

X_train_s = scaler.transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
slr = ScratchLogisticRegression(num_iter=500, lr=0.01,C=0.01,  verbose=False)
slr.fit(X_train_s, y_train, X_test_s, y_test)
pre1 = slr.predict(X_test_s)
print(pre1.shape)
```

```python
slr.predict_proba(X_test_s)
```

```python
# 評価
print("Accuracy（正解率）: ", accuracy_score(y_test, pre1))
print("Precision（適合率）: ", precision_score(y_test, pre1))
print("Recall（再現率）: ", recall_score(y_test, pre1))
print("F値: ", f1_score(y_test, pre1))
```

```python
# scikit-learnを使用したロジスティック回帰
from sklearn.linear_model import LogisticRegression

sk_logi = LogisticRegression() 
sk_logi.fit(X_train_s, y_train)
pre2 = sk_logi.predict(X_test_s)
print(pre2.shape)
```

```python
# 評価
print("Accuracy（正解率）: ", accuracy_score(y_test, pre2))
print("Precision（適合率）: ", precision_score(y_test, pre2))
print("Recall（再現率）: ", recall_score(y_test, pre2))
print("F値: ", f1_score(y_test, pre2))
```

# 問題7　決定領域の可視化

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches
import numpy as np
```

```python
def decision_region(X, y, model, step=0.01, title='decision region', xlabel='xlabel', ylabel='ylabel', target_names=['versicolor', 'virginica']):
    """
    2値分類を2次元の特徴量で学習したモデルの決定領域を描く。
    背景の色が学習したモデルによる推定値から描画される。
    散布図の点は訓練データまたは検証データである。

    Parameters
    ----------------
    X : ndarray, shape(n_samples, 2)
        特徴量
    y : ndarray, shape(n_samples,)
        ラベル
    model : object
        学習したモデルのインスンタスを入れる
    step : float, (default : 0.1)
        推定値を計算する間隔を設定する
    title : str
        グラフのタイトルの文章を与える
    xlabel, ylabel : str
        軸ラベルの文章を与える
    target_names= : list of str
        凡例の一覧を与える
    """
    # setting
    scatter_color = ['red', 'blue']
    contourf_color = ['pink', 'skyblue']
    n_class = 2

    # pred
    mesh_f0, mesh_f1  = np.meshgrid(np.arange(np.min(X[:,0])-0.5, np.max(X[:,0])+0.5, step), np.arange(np.min(X[:,1])-0.5, np.max(X[:,1])+0.5, step))
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    y_pred = model.predict(mesh).reshape(mesh_f0.shape)

    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(mesh_f0, mesh_f1, y_pred, n_class-1, cmap=ListedColormap(contourf_color))
    plt.contour(mesh_f0, mesh_f1, y_pred, n_class-1, colors='y', linewidths=3, alpha=0.5)
    for i, target in enumerate(np.unique(y)):
        plt.scatter(X[y==target][:, 0], X[y==target][:, 1], s=80, color=scatter_color[i], label=target_names[i], marker='o')
    patches = [mpatches.Patch(color=scatter_color[i], label=target_names[i]) for i in range(n_class)]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()
```

```python
fit,ax = plt.subplots(figsize=(16,9))
decision_region(X_train_s, y_train, slr, step=0.01, title='logistic', xlabel='sepal_length', ylabel='petal_length', target_names=['versicolor', 'virginica'])
```

```python
fit,ax = plt.subplots(figsize=(16,9))
decision_region(X_test_s, y_test, slr, step=0.01, title='logistic', xlabel='sepal_length', ylabel='petal_length', target_names=['versicolor', 'virginica'])
```
