```python
import pandas as pd
from sklearn.datasets import load_iris
import itertools

import matplotlib.pyplot as plt
import seaborn as sns
```

# 問題1　データの取得

```python
data = load_iris()
print(type(data))
data
```

```python
X = pd.DataFrame(data=data.data,columns=['sepal_length','sepal_width', 'petal_length', 'petal_width'])
display(X)
```

```python
y = pd.DataFrame(data=data.target, columns=['Species'])
display(y)
```

# 問題2　データ結合

```python
df = pd.concat([X, y], axis=1)
df
```

# 問題3　データの確認

```python
## 先頭から4番目までのサンプルを表示させ、各特徴量が問う言ったものかを見る
df.head(4)
```

```python
## ラベルごとにサンプル数の合計を出力する
count_sum = 0
for i in df['Species'].unique():
    count_sum = len(df[df['Species']==i])
    print('ラベル{}のサンプル数の合計は{}'.format(i, count_sum))
```

```python
## 特徴量に欠損値があるかの確認
print(df.info())

df[['sepal_length','sepal_width', 'petal_length', 'petal_width']].isnull().sum(axis=0)
```

```python
## 特徴量の平均値と標準偏差、四分位数を一度に表示させる。
df[['sepal_length','sepal_width', 'petal_length', 'petal_width']].describe()
```

# 問題6

```python
# ラベルごとのサンプル数を円グラフにする
label_y = df['Species'].value_counts()
plt.pie(label_y, labels=['setosa', 'versicolor', 'virginica'], startangle=90, autopct="%1.1f%%")
plt.show()
```

```python
# 特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する
labels = ['sepal_length','sepal_width', 'petal_length', 'petal_width']
for i in labels:
    se_0 = df[df['Species']==0][i]
    se_1 = df[df['Species']==1][i]
    se_2 = df[df['Species']==2][i]
    plt.title(i)
    plt.boxplot((se_0,se_1,se_2), labels=['setosa', 'versicolor', 'virginica'])
    plt.show()
```

```python
# 特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する

labels = ['sepal_length','sepal_width', 'petal_length', 'petal_width']
for i in labels:
    se_0 = df[df['Species']==0][i]
    se_1 = df[df['Species']==1][i]
    se_2 = df[df['Species']==2][i]
    plt.title(i)
    plt.violinplot([se_0,se_1,se_2])
    plt.xticks([1,2,3],labels=['setosa', 'versicolor', 'virginica'])
    plt.show()
```

## 箱ひげ図とバイオリン図の違いと利点
それぞれの図の違いは、取得出来る情報量に差がある。
それぞれの利点は、箱ひげ図の利点は四分位数や中央値が一目でわかる点、バイオリン図の利点は値によるデータ量（データの分布）が一目でわかる点があげられる。


# 問題7

```python
sns.pairplot(df, hue="Species")
plt.show()
```

```python
labels = ['sepal_length','sepal_width', 'petal_length', 'petal_width']
for pair in itertools.combinations(labels, 2):
    df_tmp = df.loc[:, [pair[0], pair[1]]]
    display(df_tmp.corr())
```

```python
labels = ['sepal_length','sepal_width', 'petal_length', 'petal_width']
df_corr = df[labels].corr(method='pearson')
display(df_corr)
```

```python
sns.heatmap(df_corr)
plt.show()
```
