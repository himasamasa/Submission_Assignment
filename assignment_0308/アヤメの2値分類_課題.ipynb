### 4種類の指標値の意味
・Accuracy（正解率）：予測のうち、正解した予測の割合。偏りが大きいデータを扱うのには不向き。
・Precision（適合率）：陽性と予測したもののうち、実際に陽性であるものの割合。偽陰性が多い場合は用いない方がよい。
・Recall（再現率）：実際に要請であるもののうち、正しく陽性と予測できたものの割合。誤って陰性と判断しては困る場合に用いられる。偽陰性が多いと困る場合には用いない方がよい。
・F値：対照的な特徴を持つ適合率と再現率の調和平均。偽陽性や偽陰性が極端に多い場合には、値が小さくなるため、どちらの指標もそれなりに高い値である必要がある。

# 問題7

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches
import numpy as np
```

```python
def decision_region(X, y, model, step=0.01, title='decision region', xlabel='xlabel', ylabel='ylabel', target_names=['versicolor', 'virginica']):
    """
    2値分類を2次元の特徴量で学習したモデルの決定領域を描く。
    背景の色が学習したモデルによる推定値から描画される。
    散布図の点は訓練データまたは検証データである。

    Parameters
    ----------------
    X : ndarray, shape(n_samples, 2)
        特徴量
    y : ndarray, shape(n_samples,)
        ラベル
    model : object
        学習したモデルのインスンタスを入れる
    step : float, (default : 0.1)
        推定値を計算する間隔を設定する
    title : str
        グラフのタイトルの文章を与える
    xlabel, ylabel : str
        軸ラベルの文章を与える
    target_names= : list of str
        凡例の一覧を与える
    """
    # setting
    scatter_color = ['red', 'blue']
    contourf_color = ['pink', 'skyblue']
    n_class = 2

    # pred
    mesh_f0, mesh_f1  = np.meshgrid(np.arange(np.min(X[:,0])-0.5, np.max(X[:,0])+0.5, step), np.arange(np.min(X[:,1])-0.5, np.max(X[:,1])+0.5, step))
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    y_pred = model.predict(mesh).reshape(mesh_f0.shape)

    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(mesh_f0, mesh_f1, y_pred, n_class-1, cmap=ListedColormap(contourf_color))
    plt.contour(mesh_f0, mesh_f1, y_pred, n_class-1, colors='y', linewidths=3, alpha=0.5)
    for i, target in enumerate(np.unique(y)):
        plt.scatter(X[y==target][:, 0], X[y==target][:, 1], s=80, color=scatter_color[i], label=target_names[i], marker='o')
    patches = [mpatches.Patch(color=scatter_color[i], label=target_names[i]) for i in range(n_class)]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()
```

# 問題10

```python
# すべての目的変数
data = load_iris()
X = pd.DataFrame(data=data.data,columns=['sepal_length','sepal_width', 'petal_length', 'petal_width']).loc[:,['sepal_length','petal_length']]
y = pd.DataFrame(data=data.target, columns=['Species'])
df = pd.concat([X, y], axis=1)

X = df[['sepal_length', 'petal_length']].values
y = df['Species'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123)

scaler = StandardScaler()
scaler.fit(X_train)
X_train_s = scaler.transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
def decision_region_2(X, y, model, step=0.01, title='decision region', xlabel='xlabel', ylabel='ylabel', target_names=['setosa', 'versicolor', 'virginica']):
    """
    Parameters
    ----------------
    X : ndarray, shape(n_samples, 2)
        特徴量
    y : ndarray, shape(n_samples,)
        ラベル
    model : object
        学習したモデルのインスンタスを入れる
    step : float, (default : 0.1)
        推定値を計算する間隔を設定する
    title : str
        グラフのタイトルの文章を与える
    xlabel, ylabel : str
        軸ラベルの文章を与える
    target_names= : list of str
        凡例の一覧を与える
    """
    # setting
    scatter_color = ['red', 'blue', 'green']
    contourf_color = ['pink', 'skyblue', 'lawngreen']
    n_class = 3

    # pred
    mesh_f0, mesh_f1  = np.meshgrid(np.arange(np.min(X[:,0])-0.5, np.max(X[:,0])+0.5, step), np.arange(np.min(X[:,1])-0.5, np.max(X[:,1])+0.5, step))
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    y_pred = model.predict(mesh).reshape(mesh_f0.shape)

    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(mesh_f0, mesh_f1, y_pred, n_class-1, cmap=ListedColormap(contourf_color))
    plt.contour(mesh_f0, mesh_f1, y_pred, n_class-1, colors='y', linewidths=3, alpha=0.5)
    for i, target in enumerate(np.unique(y)):
        plt.scatter(X[y==target][:, 0], X[y==target][:, 1], s=80, color=scatter_color[i], label=target_names[i], marker='o')
    patches = [mpatches.Patch(color=scatter_color[i], label=target_names[i]) for i in range(n_class)]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()
```

```python
# 再近傍法
neigh = KNeighborsClassifier(n_neighbors=3)
# 学習
neigh.fit(X_train_s, y_train)
# 推定
pred = neigh.predict(X_test_s)

from sklearn.metrics import precision_recall_fscore_support

# 評価
# Accuracy（正解率）
print("Accuracy：",accuracy_score(y_test, pred))
# マクロ平均
print("マクロ平均：",precision_recall_fscore_support(y_test, pred, average="macro"))
# マイクロ平均
print("マイクロ平均：",precision_recall_fscore_support(y_test, pred, average="micro"))
# 混同行列
display("混同行列：",confusion_matrix(y_test, pred))

decision_region_2(X_test_s, y_test, neigh, step=0.01, title='3-nn', xlabel='sepal_length', ylabel='petal_length', target_names=['setosa', 'versicolor', 'virginica'])
```

```python
# ロジスティック回帰
clf = LogisticRegression(random_state=0)
# 学習
clf.fit(X_train_s, y_train)
# 推定
pred = clf.predict(X_test_s)

# 評価
# Accuracy（正解率）
print("Accuracy：",accuracy_score(y_test, pred))
# マクロ平均
print("マクロ平均：",precision_recall_fscore_support(y_test, pred, average="macro"))
# マイクロ平均
print("マイクロ平均：",precision_recall_fscore_support(y_test, pred, average="micro"))
# 混同行列
display("混同行列：",confusion_matrix(y_test, pred))

decision_region_2(X_test_s, y_test, clf, step=0.01, title='LogisticRegression', xlabel='sepal_length', ylabel='petal_length', target_names=['setosa', 'versicolor', 'virginica'])
```

```python
# SVC
clf = SVC(gamma='auto')
# 学習
clf.fit(X_train_s, y_train)
# 推定
pred = clf.predict(X_test_s)

# 評価
# Accuracy（正解率）
print("Accuracy：",accuracy_score(y_test, pred))
# マクロ平均
print("マクロ平均：",precision_recall_fscore_support(y_test, pred, average="macro"))
# マイクロ平均
print("マイクロ平均：",precision_recall_fscore_support(y_test, pred, average="micro"))
# 混同行列
display("混同行列：",confusion_matrix(y_test, pred))

decision_region_2(X_test_s, y_test, clf, step=0.01, title='LogisticRegression', xlabel='sepal_length', ylabel='petal_length', target_names=['setosa', 'versicolor', 'virginica'])
```
